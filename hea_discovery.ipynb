# HEA Catalyst Screening Pipeline
### Stage 1: Composition Generator
From a pool of ~30 transition metals, generates 100k compounds with 4, 5, or 6 elements and random at%'s that add up to 100%. Then, concatenate the table into a one-column-wide vector.

```python
# The ~30 most common transition metals
# We're excluding precious metals (Pt, Ir, Ru, Rh, Pd) intentionally
# to keep compositions earth-abundant and low-cost

transition_metals = [
    'Ti', 'V',  'Cr', 'Mn', 'Fe', 'Co', 'Ni', 'Cu', 'Zn',
    'Zr', 'Nb', 'Mo', 'Tc', 'Ag', 'Hf', 'Ta', 'W',  'Re',
    'Al', 'Si', 'P',  'S',  'Ga', 'Ge', 'Sn', 'Bi', 'Sb',
    'Sc', 'Y',  'Mn'
]

print(f"Element pool size: {len(transition_metals)} elements")
```

```python
import numpy as np

def generate_random_composition(element_pool, min_elements=4, max_elements=6):
    # Step 1: pick how many elements (4, 5, or 6)
    n = np.random.randint(min_elements, max_elements + 1)

    # Step 2: pick n elements randomly from the pool
    elements = np.random.choice(element_pool, n, replace=False)

    # Step 3: generate percentages that sum to 100
    percentages = np.random.dirichlet(np.ones(n)) * 100
    percentages = np.round(percentages)

    return dict(zip(elements, percentages))

# Test it
print(generate_random_composition(transition_metals))
```

```python
import pandas as pd

n_samples = 100000

compositions = [generate_random_composition(transition_metals) for _ in range(n_samples)]
df = pd.DataFrame(compositions)

def row_to_formula(row):
    return ''.join([f"{el}{val}" for el, val in row.items() if pd.notna(val)])

df['formula'] = df.apply(row_to_formula, axis=1)
print(df['formula'].head())
```
